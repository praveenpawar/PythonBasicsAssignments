```python
4. What is the difference between a function and a function call?
Ans: A function and a function call are two distinct concepts in programming.
    
A function is a self-contained block of code that performs a specific task, 
and can be called from other parts of a program. When you define a function, 
you give it a name and specify the input parameters that it accepts (if any), 
as well as the code that it will execute when called. Functions can return values or 
perform operations without returning anything.

A function call, on the other hand, is the actual act of invoking or executing a function in your code. 
When you call a function, you use its name followed by parentheses, optionally passing in any arguments that it requires. 
The function call causes the code inside the function to be executed with the provided arguments, 
and can produce a return value if the function is designed to do so.

In summary, a function is a reusable block of code that performs a specific task, 
while a function call is the act of invoking that code at a specific point in your program. 
```

```python
9. How do you make a function variable refer to the global variable?
Ans: In Python, if you want to access a global variable from within a function, 
    you can use the global keyword to declare the variable as global.

When you declare a variable as global, any references to that variable within the 
function will refer to the global variable rather than a local variable with the same name.

Heres an example of how to use the global keyword to refer to a global variable from within a function:
    x = 10  # global variable

def my_function():
    global x   # declare x as global
    x = 20     # modify the global variable
    print("The value of x inside the function is:", x)

my_function()
print("The value of x outside the function is:", x)

In this example, the global keyword is used to declare the variable x 
as a global variable inside the my_function function. When the function is called, 
the value of x is set to 20, which modifies the global variable.

When the function completes execution, the value of x is still 20, 
as the global variable has been modified by the function. 
The second print statement outputs the value of x as 20, w
hich demonstrates that the global variable was indeed modified by the function.
```
